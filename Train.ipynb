```python
def read(im_or):
#   im_or=cv.imread(imgpath)
  gray = cv.cvtColor(im_or, cv.COLOR_BGR2GRAY)
  # claheCorrecttedFrame = clahefilter.apply(hs[:,:,2])
  thresh_98,im_1 = cv.threshold(gray, 100, 185, cv.THRESH_OTSU)
  
  # cv.imshow("screen",im_1)
  im_2=cv.copyMakeBorder(src=im_1, top=15, bottom=15, left=15, right=15, borderType=cv.BORDER_CONSTANT,value=0)
  contours_98, hierarchy_98  = cv.findContours(im_2, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)
  # cv.drawContours(im_2,contours_98,-1, (255,0,75), 2)
  plt.imshow(im_2)
  t=0
  ans=''
  while(t!=-1):
    ans+=checkdigit(contours_98[t])
    t=hierarchy_98[0][t][0]
  return ans

def checkdigit(cnt):
  plt.imshow(cv.drawContours(np.zeros(shape=(128,128)),cnt,-1, (255,0,75), 2)) 
  min=10000000
  epsilon = 0.01*cv.arcLength(cnt,True)
  approx = cv.approxPolyDP(cnt,epsilon,True)
  if cv.contourArea(cnt)<10:return ''
  for i in range(10):
    ret = cv.matchShapes(cnt_num_2[i],approx,cv.CONTOURS_MATCH_I2,0.0)
    if(ret<min):
      min=ret
      number=i
  return str(number)
```

# HueMomentsShapeMatching

```python
DataList=[]
valList=[]
for i in range(10):
    # print(os.listdir(f"dataset_of_screen/{i}"))
    for imgpath in os.listdir(f"dataset_of_screen/{i}"):
        im_or=cv.imread(f"dataset_of_screen/{i}/"+imgpath)
        # print(im_or.shape)
        gray = cv.cvtColor(im_or, cv.COLOR_BGR2GRAY)
        thresh_98,im_1 = cv.threshold(gray, 100, 185, cv.THRESH_OTSU)
        im_2=cv.copyMakeBorder(src=im_1, top=15, bottom=15, left=15, right=15, borderType=cv.BORDER_CONSTANT,value=0)
        contours_98, hierarchy_98  = cv.findContours(im_2, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)
        Dict=cv.moments(contours_98[0])
        # Dict["val"]=i
        DataList.append(Dict)
        valList.append(i)
```

```python
import pandas as pd
```

```python
Df=pd.DataFrame(DataList)
valSer=pd.Series(valList)
```

```python
Df.to_csv('moments.csv')
```

```python
valSer.to_csv('labels.csv')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Df, valSer, test_size=0.33, random_state=42)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(random_state=0)
```

```python
clf.fit(X_train,y_train)
```

```python
tree.plot_tree(clf)
```
